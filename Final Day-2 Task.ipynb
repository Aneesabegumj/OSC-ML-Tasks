```python
# Importing the necessary libraries

import pandas as pd
import numpy as np
```

```python
# Importing the Dataset

df=pd.read_csv("salary_prediction.csv")
df
```

```python
# Checking the head of the dataset

df.head(10)
```

```python
# Checking the shape of the dataset

df.shape
```

```python
# Checking the info of the dataset

df.info()
```

```python
# Checking for the null values 

df.isnull().sum()
```

```python
# Removing the null values

new_df=df.dropna()
new_df
```

```python
# Checking the shape of the new dataset

new_df.shape
```

```python
#Splitting the dataset into Independent(X) and Dependent(y) Variables

x =new_df.iloc[:, :-1].values
y =new_df.iloc[ :, -1].values
```

```python
# Independent variable(Years of Experience)

x
```

```python
# Dependent variable(Salary)

y
```

```python
# Plotting scatter plot to analyze the relationship between Salary and Years of Experience

import matplotlib.pyplot as plt
scatter=plt.scatter(x,y,color="purple")
scatter
```

```python
# Splitting the dataset for training set and testing set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 0)
```

```python
# Fitting Linear regression model to the training set

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# Predicting the Test set results

y_pred=regressor.predict(x_test)
```

```python
# Displaying y_predict(Predicted Salary)

y_pred
```

```python
# Displaying y_test(Actual Salary)

y_test
```

```python
# Checking accuracy of the model

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test,y_pred))
r2 = r2_score(y_test,y_pred)                            

print("RMSE =", rmse)
print("R2 Score=",r2)
```

Therefore, our model gives 89% accurate score
